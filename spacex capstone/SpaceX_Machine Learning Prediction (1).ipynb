This function is to plot the confusion matrix.


```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed']) 
    plt.show() 
```

```python
# compare model using cross validation 
def cross_validation(X, y, models, parameters):

    for i, model in enumerate(models):
        model = GridSearchCV(models[i], parameters[i], cv=10)
        
        # Define the number of "splits" (similar to folds)
        n_splits = 5
        random_states = [42, 100, 200, 300, 400]  # Different random states for different splits
        
        accuracy_scores = []
        split_numbers = []
        
        for j in range(n_splits):
            # Perform a single train-test split
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_states[j])
        
            # Train the model on the training set
            model.fit(X_train, y_train)
        
            # Make predictions on the test set
            y_pred = model.predict(X_test)
        
            # Calculate the accuracy score for this split
            accuracy = accuracy_score(y_test, y_pred)
            accuracy_scores.append(accuracy)
            split_numbers.append(j + 1)
        
        # Create the plot
        plt.plot(split_numbers, accuracy_scores, marker='o', linestyle='-', color='green')
        
    plt.title('Accuracy Score for Each Random Train-Test Split')
    plt.xlabel('Split Number')
    plt.ylabel('Accuracy Score')
    plt.xticks(split_numbers)
    plt.grid(True)
    plt.ylim(0, 1.05)
    plt.show()


#logistic
parameters =[{'C':[0.01,0.1,1],
             'penalty':['l2'],
             'solver':['lbfgs']},{'kernel':('linear', 'rbf','poly','rbf', 'sigmoid'),
              'C': np.logspace(-3, 3, 5),
              'gamma':np.logspace(-3, 3, 5)}]


models = [LogisticRegression(), SVC()]


cross_validation(X,Y, models, parameters)
```
